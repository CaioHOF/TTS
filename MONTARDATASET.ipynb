```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
df = pd.read_csv('/content/Links_Cadeiras_novo.csv')

l = []
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_product_info(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        product_info = {}

        dimensions_table = soup.find('table', class_='cav--c-gusfms-gjkaP-striped-odd')
        image_element = soup.find('img')
        #cor = soup.find('span', class_="cav--c-gNPphv cav--c-gNPphv-esGUUC-textStyle-bodySmallBold" )
        #cor = str(cor)[73:]
        #cor = str(cor)[:5]
        nome = soup.find('h1', class_="cav--c-fpAEqe" )
        nome = str(nome)[26:]
        nome = str(nome)[:53]
        product_info['nome'] = nome
        if dimensions_table:

            dimensions_rows = dimensions_table.find_all('tr')
            for row in dimensions_rows:
                cells = row.find_all('td')
                if len(cells) == 2:
                    attribute = cells[0].text.strip()
                    value = cells[1].text.strip()

                    if 'Altura' in attribute:
                        product_info['altura'] = value
                        product_info['altura'] = product_info['altura'].replace("cm", "")
                    elif 'Largura' in attribute:
                        product_info['largura'] = value
                        product_info['largura'] = product_info['largura'].replace("cm", "")
                    elif 'Profundidade' in attribute:
                        product_info['profundidade'] = value
                        product_info['profundidade'] = product_info['profundidade'].replace("cm", "")
                    elif 'Peso' in attribute:
                        product_info['peso'] = value
                        product_info['peso'] = product_info['peso'].replace("kg", "")
        #product_info['cor'] = cor

        if image_element:
            image_url = image_element['src']
            product_info['image_url'] = image_url


        return product_info
    else:
        return None
```

```python
#for x in range(len(df)):
for x in range(len(df['links'])):
  product = scrape_product_info(df['links'][x])
  l.append(product)

'''
  if product:
      print('Nome:', product.get('nome'))
      print('Altura:', product.get('altura'))
      print('Largura:', product.get('largura'))
      print('Profundidade:', product.get('profundidade'))
      print('URL da Imagem:', product.get('image_url'))
      print('Cor predominante:', product.get('cor'))
  else:
      print('Erro ao acessar a página ou informações não encontradas.')
'''
```
